```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
r = requests.get("https://www.century21.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/?")
c = r.content
```

```python
soup=bs(c,'html.parser')
print(soup.prettify())
```

```python
tmp = soup.find_all('div',{"class":"property-card-primary-info"})
tmp
```

```python
l = []
for i in tmp:
    d={}
    d["typesale"] = i.find("div",{"class":"pdp-listing-type sale"}).text.strip()
    #print(typesale.strip())
    d["price"] = i.find("a",{"class":"listing-price"}).text.strip()
    #print(price.strip())
    d["address1"] = i.find("div",{"class":"property-address"}).text.strip()
    #print(address1.strip())
    d["address2"] = i.find("div",{"class":"property-city"}).text.strip()
    #print(address2.strip())
    try:
        d["beds"] = i.find("div",{"class":"property-beds"}).find("strong").text.strip()
        #print(beds)
    except:
        #print("None")
        d["beds"] = None
    try:
        d["baths"] = i.find("div",{"class":"property-baths"},"strong").find("strong").text.strip()
        #print(baths)
    except:
        #print("None")
        d["baths"] = None
    try:
        d["sqft"] = i.find("div",{"class":"property-sqft"},"strong").find("strong").text.strip()
        #print(sqft)
    except:
        #print("None")
        d["sqft"] = None
    l.append(d)
```

```python
len(l)
```

```python
df = pd.DataFrame(l)
df
df.to_csv("outputcentury21.csv")
```
