## Load page for easy debugging

```python
import requests
```

```python
temp_response = requests.get('https://www.bigc.co.th/food.html')
```

```python
temp_response.text
```

## ทดลองเอาข้อมูลใส่ตัวแปรปกติของ Scrapy

```python
from scrapy.http import HtmlResponse
```

```python
response = HtmlResponse(url=temp_response.url,body=temp_response.text,encoding=temp_response.encoding)
```

## เริ่มหาค่าที่ต้องการด้วย xpath

```python
for product in response.xpath('''//div[@class='product-info']'''):
    name = product.xpath('''./h2[@class='product-name']/a/text()''').extract_first()
    prices = product.xpath('''.//span[@class='price']/text()''').re(r'[0-9.]+')[-1]
    print(name,prices)
```

```python
next_page_url = response.xpath('''//a[@class='next i-next']/@href''').extract_first()
print(next_page_url)
```

## เขียนลง File Python ปกติ พร้อมรัน

```python
%%writefile ST_BigCBot.py
import scrapy

class ieieieieiei(scrapy.Spider):
    name = 'kekekekekeke'
    start_urls = [
        'https://www.bigc.co.th/food.html',
    ]
    
    def parse(self, response):
        for product in response.xpath('''//div[@class='product-info']'''):
            name = product.xpath('''./h2[@class='product-name']/a/text()''').extract_first()
            price = product.xpath('''.//span[@class='price']/text()''').re(r'[0-9.]+')[-1]
            output = {
                'product': name,
                'price': price,
            }
            yield output
        next_page_url = response.xpath('''//a[@class='next i-next']/@href''').extract_first()
        if next_page_url is not None:
            yield scrapy.Request(response.urljoin(next_page_url))
```
