```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
url = "https://www.bigc.co.th/food.html"
response = requests.get(url,verify = False)
htmlText = response.content.decode('UTF-8')
soup = bs(htmlText,'html.parser')
fuck = soup.prettify()
productlist = soup.find_all('div',{'class':'product-info'})
productlist
```

```python
l = []
x = []
l.append([])
count = 0
for i in productlist:
    l.append([])
    fuckdict = {}
    name = i.find('h2').text
    fuckdict["Name"] = name
    try:
        price = i.find_all('span',{'class':'price'})[1].text.strip()
    except:
        price = i.find_all('span',{'class':'price'})[0].text.strip()
    fuckdict["Price"] = price
    x.append(fuckdict)
    count+=1
    
```
